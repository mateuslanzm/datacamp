```python
import pandas as pd
import numpy as np
```

```python
# Define the number of rows and columns
rows = 10
cols = 6

# Generate random integers between 0 and 100
data = np.random.randint(0, 100, size=(rows, cols))

# Create column names: col1, col2, ..., col6
column_names = [f'col{i+1}' for i in range(cols)]

# Create the DataFrame
df = pd.DataFrame(data, columns=column_names)

print(df)
```
